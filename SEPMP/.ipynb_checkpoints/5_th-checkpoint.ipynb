```python
import numpy as np
```

```python
x = np.linspace(-12, 12, 1000)
y = np.sinc(x / np.pi)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)/x', color='dodgerblue', linewidth=2)

plt.xlim(-12, 12)
plt.ylim(-0.5, 1.2)

plt.xticks(np.arange(-12, 13, 1), rotation=45)
plt.yticks(np.arange(-0.5, 1.3, 0.1), rotation=45)

plt.grid(which='both', linestyle=':', linewidth=0.7, color = 'black')

plt.gca().set_facecolor('lightpink')

plt.title('График функции sin(x)/x', fontsize=16)
plt.xlabel('Значение x', fontsize=12)
plt.ylabel('sin(x)/x', fontsize=12)
plt.legend()
plt.show()
```

# Аппроксимация

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2.1, 4.0, 6.1, 8.2, 10.1])
dy = np.array([0.2, 0.5, 0.3, 0.2, 0.5])

a = 2.01
b = 0.08

x_line = np.linspace(min(x), max(x), 100)
y_line = a * x_line + b

plt.errorbar(x, y, yerr=dy, fmt='o', label='Данные с погрешностями')
plt.plot(x_line, y_line, 'r-', label=f' y={a:.2f}x+{b:.2f}')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(-5, 5, 30)
y = 2*x**2 + 3*x + 5 + np.random.normal(0, 5, size=len(x))  # добавляем шум

# Аппроксимация полиномом 2-й степени
coeffs = np.polyfit(x, y, deg=2)
poly_func = np.poly1d(coeffs) #превращает найденные коэффициенты в полиномиальную функцию, которую можно использовать как обычную

# Генерация линии аппроксимации
x_fit = np.linspace(min(x), max(x), 200)
y_fit = poly_func(x_fit)

plt.scatter(x, y, label="Данные", color="blue")
plt.plot(x_fit, y_fit, label=f"Аппроксимация: {coeffs[0]:.2f}x² + {coeffs[1]:.2f}x + {coeffs[2]:.2f}", color="red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Сгенерированные данные и аппроксимация")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(-5, 5, 30)
y = 2*x**2 + 3*x + 5 + np.random.normal(0, 5, size=len(x)) 

# Аппроксимация полиномом 1-й степени (прямая)
coeffs = np.polyfit(x, y, deg=1)
poly_func = np.poly1d(coeffs)

# Генерация линии аппроксимации
x_fit = np.linspace(min(x), max(x), 200)
y_fit = poly_func(x_fit)

plt.scatter(x, y, label="Данные", color="blue")
plt.plot(x_fit, y_fit, label=f"Аппроксимация: {coeffs[0]:.2f}x + {coeffs[1]:.2f}", color="red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Линейная аппроксимация данных")
plt.grid(True)
plt.show()
```

## Объединение

```python
animals_1 = pd.DataFrame({'вид': ['собака', 'кошка', 'мышь', 'слон'],
                         'средняя продолжительность жизни': ['10-13 лет', '12-17 лет', '1-2 года', '50-70 лет']})

animals_2 = pd.DataFrame({'вид': ['собака', 'кошка', 'мышь', 'лев'],
                         'средний вес': ['10-30 кг', '4-6 кг', '0,02-0,05 кг', '150-250 кг']})

merged_inner = pd.merge(animals_1, animals_2, on='вид', how='inner')


print(merged_inner)
```

```python
merged_inner = merged_inner.sort_values(by="вид", ascending= True)


print(merged_inner)
```

```python
merged_right = pd.merge(animals_1, animals_2, on='вид', how='right')

print(merged_right)
```

```python
merged_left = pd.merge(animals_1, animals_2, on='вид', how='left')

print(merged_left)
```

```python
merged_outer = pd.merge(animals_1, animals_2, on='вид', how='outer')

print(merged_outer)
```

```python
Capitals = {'Russia': 'Moscow', 'China': 'Beijing', 'USA': 'Washington'}
Capitals2 = dict(Russia = 'Moscow', China = 'Beijing', USA = 'Washington')
Capitals3 = dict([("Russia", "Moscow"), ("China", "Beijing"), ("USA", "Washington")])
Capitals4 = dict(zip(["Russia", "China", "USA"], ["Moscow", "Beijing", "Washington"]))

print(Capitals)
print(Capitals2)
print(Capitals3)
print(Capitals4)
```

```python
print('В какой стране вы живете?')
country = input()

if country in Capitals:
    print('Столица вашей страны', Capitals[country])
else:
    print('Как называется столица вашей страны?')
    city = input()
    Capitals[country] = city
```
