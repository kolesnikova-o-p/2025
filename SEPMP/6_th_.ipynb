# Лямбда

Лямбда-функции могут иметь любое количество аргументов, но у каждой может быть только одно выражение. Выражение вычисляется и возвращается. Эти функции могут быть использованы везде, где требуется объект-функция.

Если функцию нужно вызывать многократно, следует написать обычную функцию def

```python
double = lambda x: x*2
```

```python
def double(x):
		return x * 2
```

```python
def defined_cube(y):
    return y*y*y


lambda_cube = lambda y: y*y*y

print(defined_cube(2))
print(lambda_cube(2))
```

Без использования лямбды: Здесь обе функции возвращают заданное значение, возведенное в куб. Но при использовании def, нам пришлось определить функцию с именем defined_cube() дать ей входную величину.  После выполнения нам также понадобилось возвратить результат, из того места, откуда была вызвана функция, и мы сделали это, используя ключевое слово return.

С применением лямбды: Определение лямбды не включает оператор return, а всегда содержит возвращенное выражение. Мы также можем поместить определение лямбды в любое место, где ожидается функция, и нам не нужно присваивать его переменной.

```python
my_list = [1, 3, 4, 6, 10, 11, 15, 12, 14]
new_list = list(filter(lambda x: (x%2 == 0) , my_list))
print(new_list)
```

# Глобальные и локальные переменные

Когда вы используете переменную внутри функции, Python пытается найти эту переменную в локальной области функции. Если переменная не найдена там, то  он ищет её в глобальной области.

Если вы хотите изменить значение глобальной переменной внутри функции, вы должны явно указать, что переменная является глобальной, используя ключевое слово global. Если вы этого не сделаете, Python будет считать, что вы хотите создать новую локальную переменную с тем же именем.


```python
def loc():
    y = 10  
    print(y)

loc() 
print(y)  
```

```python
z = 10  

def glob():
    print(z)  

glob()  
print(z)  
```

```python
a = 10  

def example():
    a = 20  
    print(a) 

example() 
print(a)  
```

```python
x = 10  

def example2():
    global x  
    x = 20  
    print(x)  

example2()  
print(x) 
```

```python
'''
Декораторы — это, по сути, "обёртки", которые дают нам возможность изменить поведение функции, не изменяя её код. 

Любая достаточно общая функциональность, которую можно «прикрепить» к поведению существующего класса или функции, 
является отличным примером использования декораторов. Сюда входит:
обеспечение контроля доступа и аутентификации,
инструментарий и функции управления временем,
ограничение скорости,
кэширование и многое другое.
'''
def null_decorator(func):
    return func
def greet():
    return 'Hello!'

greet = null_decorator(greet)

greet()
```

```python
@null_decorator
def greet():
    return 'Hello!'
greet()
```

```python
def uppercase(func):
    def wrapper():
        original_result = func()
        modified_result = original_result.upper()
        return modified_result
    return wrapper

@uppercase
def greet():
    return 'Hello!'

greet()
```

```python
from functools import wraps

def uppercase(func):
    @wraps(func)
    def wrapper():
        return func().upper()
    return wrapper

@uppercase
def greet():
    return "Hello!"

print(greet())         
print(greet.__wrapped__()) 
```

```python
import time

def timer(func):
    def wrapper(*args):
        start_time = time.time()
        result = func(*args)
        end_time = time.time()
        print(f" функция {func.__name__} затратила {end_time - start_time} секунд на выполнение")
        return result
    return wrapper

@timer
def zaderzhka():
    time.sleep(5)

zaderzhka()
```

```python
def timer(func):
    def wrapper(*args):
        start_time = time.time()
        result = func(*args)
        end_time = time.time()
        print(f"Функция {func.__name__} затратила {end_time - start_time} секунд на выполнение")
        return result
    return wrapper

@timer
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

factorial(100)
```

```python
def iterations(func):
    def wrapper(*args):
        print(f"Внимание! Вызвана функция {func.__name__} с аргументами {args}")
        result = func(*args)
        return result
    return wrapper

@iterations
def function(n):
    j=1
    for i in range(1, n+1):
        print(f"Итерация {i}:")
        j *=i
        print(f"Выполняю итерацию № {i}...")
        print(j)

function(5)
```

В коде класса Person определен конструктор - функция __init__. Конструктор должен принимать как минимум один параметр ссылку на текущий объект - self. 

| `self`      | текущий объект класса   |
| `self.name` | атрибут объекта         |
| `name_`     | просто параметр функции |

```python
class DomesticAnimal:
    def __init__(self, name, age, species):
        self.name = name
        self.age = age
        self.species = species

    def eat(self):
        print(f"{self.name} кушает.")

    def sleep(self):
        print(f"{self.name} спит.")

    def play(self):
        print(f"{self.name} играет.")

    def sound(self):
        if self.species == "собака":
            print("Гав!")
        elif self.species == "кошка":
            print("Мяу!")
        else:
            print("Неизвестный звук.")

    def info(self):
        print(f"Имя: {self.name}, Возраст: {self.age}, Вид: {self.species}")

# Создаем экземпляр класса
dog = DomesticAnimal("Шарик", 3, "собака")
cat = DomesticAnimal("Мурзик", 2, "кошка")

# Вызываем методы
dog.eat()
dog.sleep()
dog.play()
dog.sound()
dog.info()

cat.eat()
cat.sleep()
cat.play()
cat.sound()
cat.info()
```

```python
class Persons:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Привет, меня зовут {self.name} и мне {self.age} лет.")

class Student(Persons):
    def __init__(self, name, age, course):
        super().__init__(name, age) 
        self.course = course

    def study(self):
        print(f"Я изучаю {self.course}.")

ivan = Student("Ivan", 18, "Python")
ivan.greet()  
ivan.study() 
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def work(self):
        print(f"{self.name} работает и зарабатывает {self.salary} руб.")

    def take_break(self):
        print(f"{self.name} делает перерыв на 15 минут")


class Student:
    def __init__(self, name, university):
        self.name = name
        self.university = university

    def study(self):
        print(f"{self.name} учится в {self.university}")

    def take_break(self):
        print(f"{self.name} отдыхает 5 минут после учёбы")


class WorkingStudent(Employee, Student):
    def __init__(self, name, salary, university):
        Employee.__init__(self, name, salary)
        Student.__init__(self, name, university)

    def work(self):
        print(f"{self.name} одновременно работает и учится")
        super().work()  

    def take_break(self):
        print(f"{self.name} комбинирует отдых с учёбой и работой")
        Employee.take_break(self)
        Student.take_break(self)


tom = WorkingStudent("Tom", 100000, "MIPT")

tom.work()       
tom.study()      
tom.take_break()
```
