# PyGame

pygame — это популярный кроссплатформенный модуль для языка Python, предназначенный для разработки 2D-игр и мультимедийных приложений. Он обёрнут вокруг низкоуровневой библиотеки SDL (Simple DirectMedia Layer) и предоставляет удобный, «питонический» интерфейс для работы с графикой, звуком, событиями и вводом.

Основные возможности

Создание окон, отрисовка спрайтов, работа с изображениями (PNG, JPG, GIF и др.).

Поддержка прозрачности, альфа-каналов, вращения, масштабирования.

Управление кадрами через объект pygame.time.Clock для плавной анимации.

Обработка ввода

События клавиатуры, мыши, геймпадов и сенсорных устройств.

Гибкая система очередей событий (pygame.event).

Воспроизведение звуковых эффектов (pygame.mixer.Sound) и музыки (pygame.mixer.music) в разных форматах.

Работа с текстом

Рендеринг шрифтов (через pygame.font) для вывода текста на экран.

```python
import pygame
pygame.init()                         
screen = pygame.display.set_mode((400, 300))  
pygame.display.set_caption("Самый простой пример")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:  
            running = False

    screen.fill((30, 144, 255))   
    
    pygame.display.flip()               

pygame.quit()                        
```

Аргумент цвета может быть:

тройкой (R, G, B) — значения красного, зелёного и синего каналов (0–255)

четвёркой (R, G, B, A) — если нужен альфа-канал (прозрачность)

### Смайл

```python
import pygame

pygame.init()                         
screen = pygame.display.set_mode((400, 300))  
pygame.display.set_caption("Весёлый смайлик")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:  
            running = False




    
    screen.fill((255, 255, 255)) 
    pygame.draw.circle(screen, (255, 230, 0), (200, 150), 100)    
    pygame.draw.circle(screen, (0, 0, 0),   (200, 150), 100, 3)     
    pygame.draw.circle(screen, (0, 0, 0), (160, 120), 15)
    pygame.draw.circle(screen, (0, 0, 0), (240, 120), 15)
    smile_rect = pygame.Rect(130, 120, 140, 100)
    pygame.draw.arc(screen, (0, 0, 0), smile_rect, 3.14, 2*3.14, 4)
    pygame.display.flip()             

pygame.quit()
```

Вариант на 3 балла

```python
import pygame
pygame.init()
font = pygame.font.SysFont(None, 24)

screen = pygame.display.set_mode((960, 768))
pygame.display.set_caption("Горы из фигур")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((250, 210, 150))
    pygame.draw.circle(screen, (255, 240, 0), (470, 120), 50)
    pygame.draw.polygon(screen, (250, 160, 50), [
        (0, 260), (150, 80), (320, 220), (460, 140),
        (600, 260), (760, 150), (952, 260), (0, 260)
    ])


    # 1) левый маленький овал
    pygame.draw.ellipse(screen, (170, 70, 50), (0, 230, 100, 240))

    # 2) большой овал слева
    pygame.draw.ellipse(screen, (170, 70, 50), (60, 180, 160, 300))

    # 3) наклонённый квадрат 
    pygame.draw.polygon(screen, (170, 70, 50), [
        (210, 400), (280, 320), (350, 400), (280, 480)
    ])

    # 4) длинный многоугольник 
    pygame.draw.polygon(screen, (170, 70, 50), [
        (320, 480), (370, 250), (460, 300), (520, 330), (560, 480)
    ])

    # 5) средний овал – «гладкая» гора
    pygame.draw.ellipse(screen, (170, 70, 50), (500, 200, 200, 260))

    # 6) правый угловатый многоугольник – последний пик
    pygame.draw.polygon(screen, (170, 70, 50), [
        (700, 480), (760, 340), (800, 390),
        (880, 300), (952, 350), (1000, 400)
    ])


    pygame.draw.rect(screen, (170, 120, 130), (0, 500, 1152, 268))


    mx, my = pygame.mouse.get_pos()
    text = font.render(f"x: {mx}, y: {my}", True, (0, 0, 0))
    screen.blit(text, (10, 10))
    pygame.draw.rect(screen, (170, 70, 50), (0, 400, 1152, 500))
    pygame.draw.rect(screen, (170, 120, 130), (0, 500, 1152, 268))

    pygame.display.flip()
    

pygame.quit()
```
