# Повторение pygame

| Метод / Функция                       | Назначение                                | Пример                                         |
| ------------------------------------- | ----------------------------------------- | ---------------------------------------------- |
| `pygame.init()`                       | Инициализирует все модули Pygame          | `pygame.init()`                                |
| `pygame.display.set_mode((w, h))`     | Создаёт окно размером `w × h`             | `screen = pygame.display.set_mode((400, 300))` |
| `pygame.display.set_caption("текст")` | Задаёт заголовок окна                     | `pygame.display.set_caption("Моя игра")`       |
| `pygame.display.flip()`               | Обновляет экран (показывает нарисованное) | `pygame.display.flip()`                        |
| `screen.fill((r,g,b))`                | Закрашивает экран цветом RGB              | `screen.fill((30,144,255))`                    |


| Метод                                                 | Назначение           | Пример                                                         |
| ----------------------------------------------------- | -------------------- | -------------------------------------------------------------- |
| `pygame.draw.circle(surface, color, center, radius)`  | Рисует круг          | `pygame.draw.circle(screen, (0,255,0), (x, y), radius)`        |
| `pygame.draw.rect(surface, color, rect)`              | Рисует прямоугольник | `pygame.draw.rect(screen, (255,0,0), (10,10,50,30))`           |
| `pygame.draw.line(surface, color, start, end, width)` | Рисует линию         | `pygame.draw.line(screen, (255,255,255), (0,0), (100,100), 2)` |


| Метод / Константа                                      | Назначение                                         | Пример                               |
| ------------------------------------------------------ | -------------------------------------------------- | ------------------------------------ |
| `pygame.event.get()`                                   | Получить все события (клавиши, мышь, выход и т.п.) | `for event in pygame.event.get():`   |
| `pygame.QUIT`                                          | Событие выхода (нажатие на крестик окна)           | `if event.type == pygame.QUIT:`      |
| `pygame.KEYDOWN`                                       | Событие — нажата клавиша                           | `elif event.type == pygame.KEYDOWN:` |
| `event.key`                                            | Код клавиши, нажатой пользователем                 | `if event.key == pygame.K_w:`        |
| `pygame.K_w`, `pygame.K_s`, `pygame.K_a`, `pygame.K_d` | Константы для клавиш `W`, `S`, `A`, `D`            | `if event.key == pygame.K_w:`        |


| Метод / Класс                         | Назначение                              | Пример                                        |
| ------------------------------------- | --------------------------------------- | --------------------------------------------- |
| `pygame.mixer.Sound("файл.wav")`      | Загружает короткий звук                 | `key_sound = pygame.mixer.Sound("click.wav")` |
| `.play()`                             | Воспроизвести звук                      | `key_sound.play()`                            |
| `pygame.mixer.music.load("файл.mp3")` | Загружает фоновую музыку                | `pygame.mixer.music.load("background.mp3")`   |
| `pygame.mixer.music.play(-1)`         | Воспроизвести музыку (`-1` — зациклить) | `pygame.mixer.music.play(-1)`                 |
| `pygame.mixer.music.pause()`          | Поставить музыку на паузу               | `pygame.mixer.music.pause()`                  |
| `pygame.mixer.music.unpause()`        | Снять с паузы                           | `pygame.mixer.music.unpause()`                |


# Музыка в pygame

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("Движение кружка с фоном и звуками")

key_sound = pygame.mixer.Sound("click.wav")         
pygame.mixer.music.load("background.mp3")           
pygame.mixer.music.play(-1)                        

x, y = 200, 150
radius = 20
step = 20

running = True
paused = False  

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:  
                if paused:
                    pygame.mixer.music.unpause()
                    paused = False
                else:
                    pygame.mixer.music.pause()
                    paused = True

            key_sound.play()

            if event.key == pygame.K_w:
                y -= step
            elif event.key == pygame.K_s:
                y += step
            elif event.key == pygame.K_a:
                x -= step
            elif event.key == pygame.K_d:
                x += step

            x = max(radius, min(400 - radius, x))
            y = max(radius, min(300 - radius, y))

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (x, y), radius)
    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("Движение кружка с фоном, паузой и громкостью")

key_sound = pygame.mixer.Sound("click.wav")
pygame.mixer.music.load("background.mp3")
pygame.mixer.music.play(-1)

music_volume = 0.5
sound_volume = 0.7
pygame.mixer.music.set_volume(music_volume)
key_sound.set_volume(sound_volume)

x, y = 200, 150
radius = 20
step = 20

running = True
paused = False

font = pygame.font.SysFont(None, 24)

def draw_slider(x, y, width, value, label):
    pygame.draw.rect(screen, (180, 180, 180), (x, y, width, 10)) 
    pos = x + int(value * width)
    pygame.draw.circle(screen, (255, 100, 0), (pos, y + 5), 7)
    text = font.render(label, True, (255, 255, 255))
    screen.blit(text, (x, y - 22))

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:
                paused = not paused
                if paused:
                    pygame.mixer.music.pause()
                else:
                    pygame.mixer.music.unpause()

            elif event.key == pygame.K_UP:     
                music_volume = min(1.0, music_volume + 0.1)
                pygame.mixer.music.set_volume(music_volume)
            elif event.key == pygame.K_DOWN:   
                music_volume = max(0.0, music_volume - 0.1)
                pygame.mixer.music.set_volume(music_volume)
            elif event.key == pygame.K_RIGHT:    
                sound_volume = min(1.0, sound_volume + 0.1)
                key_sound.set_volume(sound_volume)
            elif event.key == pygame.K_LEFT:   
                sound_volume = max(0.0, sound_volume - 0.1)
                key_sound.set_volume(sound_volume)

            if event.key in (pygame.K_w, pygame.K_a, pygame.K_s, pygame.K_d):
                key_sound.play()
                if event.key == pygame.K_w:
                    y -= step
                elif event.key == pygame.K_s:
                    y += step
                elif event.key == pygame.K_a:
                    x -= step
                elif event.key == pygame.K_d:
                    x += step

                x = max(radius, min(400 - radius, x))
                y = max(radius, min(300 - radius, y))

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (x, y), radius)

    if paused:
        pygame.draw.rect(screen, (255, 255, 255), (185, 130, 10, 40)) 
        pygame.draw.rect(screen, (255, 255, 255), (205, 130, 10, 40)) 
    else:
        pygame.draw.polygon(screen, (255, 255, 255), [(185, 130), (185, 170), (215, 150)]) 

    draw_slider(50, 260, 150, music_volume, "Музыка")
    draw_slider(250, 260, 100, sound_volume, "Звук")

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("Движение кружка с фоном, паузой и интерактивными ползунками")

key_sound = pygame.mixer.Sound("click.wav")
pygame.mixer.music.load("background.mp3")
pygame.mixer.music.play(-1)

music_volume = 0.5
sound_volume = 0.7
pygame.mixer.music.set_volume(music_volume)
key_sound.set_volume(sound_volume)

x, y = 200, 150
radius = 20
step = 20

running = True
paused = False
dragging_music = False
dragging_sound = False

font = pygame.font.SysFont(None, 24)

def draw_slider(x, y, width, value, label):
    pygame.draw.rect(screen, (180, 180, 180), (x, y, width, 10))  
    pos = x + int(value * width)
    pygame.draw.circle(screen, (255, 100, 0), (pos, y + 5), 7)
    text = font.render(label, True, (255, 255, 255))
    screen.blit(text, (x, y - 22))
    return (pos, y + 5)

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:
                paused = not paused
                if paused:
                    pygame.mixer.music.pause()
                else:
                    pygame.mixer.music.unpause()

            if event.key in (pygame.K_w, pygame.K_a, pygame.K_s, pygame.K_d):
                key_sound.play()
                if event.key == pygame.K_w:
                    y -= step
                elif event.key == pygame.K_s:
                    y += step
                elif event.key == pygame.K_a:
                    x -= step
                elif event.key == pygame.K_d:
                    x += step

                x = max(radius, min(400 - radius, x))
                y = max(radius, min(300 - radius, y))

        elif event.type == pygame.MOUSEBUTTONDOWN:
            mx, my = event.pos
            music_x, music_y = 50, 260
            sound_x, sound_y = 250, 260
            music_width = 150
            sound_width = 100

            if abs(mx - (music_x + int(music_volume * music_width))) < 10 and abs(my - (music_y + 5)) < 10:
                dragging_music = True
            elif abs(mx - (sound_x + int(sound_volume * sound_width))) < 10 and abs(my - (sound_y + 5)) < 10:
                dragging_sound = True

        elif event.type == pygame.MOUSEBUTTONUP:
            dragging_music = False
            dragging_sound = False

        elif event.type == pygame.MOUSEMOTION:
            if dragging_music:
                mx = event.pos[0]
                music_x, music_width = 50, 150
                music_volume = max(0.0, min(1.0, (mx - music_x) / music_width))
                pygame.mixer.music.set_volume(music_volume)
            elif dragging_sound:
                mx = event.pos[0]
                sound_x, sound_width = 250, 100
                sound_volume = max(0.0, min(1.0, (mx - sound_x) / sound_width))
                key_sound.set_volume(sound_volume)

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (x, y), radius)


    draw_slider(50, 260, 150, music_volume, "Музыка")
    draw_slider(250, 260, 100, sound_volume, "Звук")

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

# Изменение размера

```python
import pygame

pygame.init()

width, height = 400, 300
screen = pygame.display.set_mode((width, height), pygame.RESIZABLE)
pygame.display.set_caption("Плавное движение кружка")

x, y = width // 2, height // 2
radius = 20
speed = 0.3  

running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.VIDEORESIZE:
            width, height = event.w, event.h
            screen = pygame.display.set_mode((width, height), pygame.RESIZABLE)

    keys = pygame.key.get_pressed()
    if keys[pygame.K_UP]:
        y -= speed
    if keys[pygame.K_DOWN]:
        y += speed
    if keys[pygame.K_LEFT]:
        x -= speed
    if keys[pygame.K_RIGHT]:
        x += speed

    x = max(radius, min(width - radius, x))
    y = max(radius, min(height - radius, y))

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (int(x), int(y)), radius)
    pygame.display.flip()


pygame.quit()
```

# Игра пушка

```python
import pygame
pygame.init()

screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("Отскакивающий шарик с гравитацией")

x, y = 200, 150
radius = 20

vy = 0                 # вертикальная скорость
gravity = 0.5          # сила гравитации
bounce_loss = 0.7      # потеря энергии при отскоке (0.7 = теряет 30%)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:  
                vy = -10

    vy += gravity
    y += vy

    if y + radius >= 300:
        y = 300 - radius
        vy = -vy * bounce_loss 
        
        if abs(vy) < 1:
            vy = 0

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (int(x), int(y)), radius)
    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import math
pygame.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("Бросок в баскетбольное кольцо")

x, y = 100, 300  # стартовая позиция
radius = 15

# Физика
vx, vy = 0, 0           # скорости по осям
gravity = 0.5           # гравитация
power = 12              # сила броска
angle = 45              # угол броска в градусах
bounce_loss = 0.7       # потеря энергии при ударе
on_ground = True

# Параметры кольца
hoop_x, hoop_y = 500, 200
hoop_width, hoop_height = 60, 10

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            # Изменяем угол броска
            if event.key == pygame.K_UP:
                angle = min(angle + 5, 80)
            elif event.key == pygame.K_DOWN:
                angle = max(angle - 5, 10)
            # Бросок
            elif event.key == pygame.K_SPACE and on_ground:
                rad = math.radians(angle)
                vx = power * math.cos(rad)
                vy = -power * math.sin(rad)
                on_ground = False

    if not on_ground:
        vy += gravity
        x += vx
        y += vy

    if y + radius >= 400:
        y = 400 - radius
        vy = -vy * bounce_loss
        vx *= bounce_loss
        if abs(vy) < 1:
            vy = 0
            on_ground = True

    # Проверка попадания в кольцо
    hoop_rect = pygame.Rect(hoop_x, hoop_y, hoop_width, hoop_height)
    ball_rect = pygame.Rect(x - radius, y - radius, radius * 2, radius * 2)
    scored = ball_rect.colliderect(hoop_rect)

    screen.fill((135, 206, 235))
    pygame.draw.rect(screen, (255, 100, 0), hoop_rect) 
    pygame.draw.circle(screen, (255, 165, 0), (int(x), int(y)), radius)  

    font = pygame.font.Font(None, 24)
    text = font.render(f"Угол: {angle}°", True, (0, 0, 0))
    screen.blit(text, (10, 10))

    if scored:
        hit_text = font.render("ПОПАДАНИЕ!", True, (0, 255, 0))
        screen.blit(hit_text, (250, 50))

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import math
pygame.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("Бросок в баскетбольное кольцо")

background = pygame.image.load("basketball.jpg")
background = pygame.transform.scale(background, (600, 400))
ball_image = pygame.image.load("ball.jpg")
ball_image = pygame.transform.scale(ball_image, (30, 30))

pygame.mixer.init()
pygame.mixer.music.load("play.wav")
pygame.mixer.music.play(-1) 

score_sound = pygame.mixer.Sound("score.wav")  

start_x, start_y = 100, 300
x, y = start_x, start_y
radius = 15

vx, vy = 0, 0
gravity = 0.5
power = 12
angle = 45
bounce_loss = 0.7
on_ground = True

hoop_x, hoop_y = 500, 200
hoop_width, hoop_height = 60, 10

font = pygame.font.Font(None, 24)

scored_flag = False
score = 0

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                angle = min(angle + 5, 80)
            elif event.key == pygame.K_DOWN:
                angle = max(angle - 5, 10)
            elif event.key == pygame.K_w:
                power = min(power + 1, 25)
            elif event.key == pygame.K_s:
                power = max(power - 1, 5)
            elif event.key == pygame.K_SPACE and on_ground:
                rad = math.radians(angle)
                vx = power * math.cos(rad)
                vy = -power * math.sin(rad)
                on_ground = False

    if not on_ground:
        vy += gravity
        x += vx
        y += vy

    if y + radius >= 400:
        y = 400 - radius
        vy = -vy * bounce_loss
        vx *= bounce_loss
        if abs(vy) < 1 and abs(vx) < 1:
            x, y = start_x, start_y
            vx, vy = 0, 0
            on_ground = True
            scored_flag = False

    hoop_rect = pygame.Rect(hoop_x, hoop_y, hoop_width, hoop_height)
    ball_rect = pygame.Rect(x - radius, y - radius, radius * 2, radius * 2)

    if ball_rect.colliderect(hoop_rect) and not scored_flag:
        scored_flag = True
        score += 1
        score_sound.play()

    screen.blit(background, (0, 0))

    if on_ground:
        points = []
        rad = math.radians(angle)
        temp_vx = power * math.cos(rad)
        temp_vy = -power * math.sin(rad)
        temp_x, temp_y = start_x, start_y
        for i in range(60):
            temp_vy += gravity
            temp_x += temp_vx
            temp_y += temp_vy
            if temp_y + radius >= 400:
                break
            points.append((int(temp_x), int(temp_y)))
        if len(points) > 1:
            pygame.draw.lines(screen, (0, 0, 0), False, points, 2)

    pygame.draw.rect(screen, (255, 100, 0), hoop_rect)
    screen.blit(ball_image, (int(x - radius), int(y - radius)))

    text_angle = font.render(f"Угол: {angle}°", True, (255, 0, 0))
    text_power = font.render(f"Сила: {power}", True, (255, 0, 0))
    text_score = font.render(f"Счёт: {score}", True, (255, 0, 0))
    screen.blit(text_angle, (10, 10))
    screen.blit(text_power, (10, 30))
    screen.blit(text_score, (10, 50))

    if scored_flag:
        hit_text = font.render("ПОПАДАНИЕ!", True, (0, 255, 0))
        screen.blit(hit_text, (250, 50))

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import random

pygame.init()
screen_width, screen_height = 400, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Лягушка Jump")

gravity = 1
jump_speed = -20
y_speed = 0
x = screen_width // 2
y = screen_height // 2

background = pygame.image.load("day.jpg")
frog_image = pygame.image.load("frog2.jpg").convert()
frog_image.set_colorkey((255, 255, 255))
background = pygame.transform.scale(background, (screen_width, screen_height))
frog_image = pygame.transform.scale(frog_image, (40, 40))

platforms = [(random.randint(0, 350), random.randint(50, 550)) for _ in range(8)]
platform_width = 60
platform_height = 5

clock = pygame.time.Clock()
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    if keys[pygame.K_a]:
        x -= 5
    if keys[pygame.K_d]:
        x += 5

    # горизонтальные границы
    if x < 0:
        x = screen_width
    elif x > screen_width:
        x = 0

    y_speed += gravity
    y += y_speed

    frog_rect = pygame.Rect(
        x - frog_image.get_width() // 2,
        y - frog_image.get_height() // 2,
        frog_image.get_width(),
        frog_image.get_height()
    )

    if y > screen_height:
        y = screen_height // 2
        y_speed = 0

    if y_speed > 0:
        for px, py in platforms:
            if (frog_rect.bottom >= py and frog_rect.bottom - y_speed < py and
                px < x < px + platform_width):
                y_speed = jump_speed  
                break

    if y < screen_height // 2:
        dy = screen_height // 2 - y
        y = screen_height // 2
        platforms = [(px, py + dy) for (px, py) in platforms]
        while len(platforms) < 8:
            platforms.append((random.randint(0, 350), random.randint(-50, 0)))
        platforms = [(px, py) for (px, py) in platforms if py < screen_height]

    screen.blit(background, (0, 0))
    screen.blit(frog_image, frog_rect)

    for px, py in platforms:
        pygame.draw.line(screen, (255, 255, 255),
                         (px, py), (px + platform_width, py), platform_height)

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
