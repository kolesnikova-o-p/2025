```python
import pygame

pygame.init()

width, height = 400, 300
screen = pygame.display.set_mode((width, height), pygame.RESIZABLE)
pygame.display.set_caption("  ")

x, y = width // 2, height // 2
radius = 20
speed = 0.3  

running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.VIDEORESIZE:
            width, height = event.w, event.h
            screen = pygame.display.set_mode((width, height), pygame.RESIZABLE)

    keys = pygame.key.get_pressed()
    if keys[pygame.K_UP]:
        y -= speed
    if keys[pygame.K_DOWN]:
        y += speed
    if keys[pygame.K_LEFT]:
        x -= speed
    if keys[pygame.K_RIGHT]:
        x += speed

    x = max(radius, min(width - radius, x))
    y = max(radius, min(height - radius, y))

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (int(x), int(y)), radius)
    pygame.display.flip()


pygame.quit()
```

```python
import pygame
pygame.init()

screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("   ")

x, y = 200, 150
radius = 20

vy = 0                 #  
gravity = 0.5          #  
bounce_loss = 0.7      #     (0.7 =  30%)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:  
                vy = -10

    vy += gravity
    y += vy

    if y + radius >= 300:
        y = 300 - radius
        vy = -vy * bounce_loss 
        
        if abs(vy) < 1:
            vy = 0

    screen.fill((30, 144, 255))
    pygame.draw.circle(screen, (0, 255, 0), (int(x), int(y)), radius)
    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import math
pygame.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("   ")

x, y = 100, 300  #  
radius = 15

# 
vx, vy = 0, 0           #   
gravity = 0.5           # 
power = 12              #  
angle = 45              #    
bounce_loss = 0.7       #    
on_ground = True

#  
hoop_x, hoop_y = 500, 200
hoop_width, hoop_height = 60, 10

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            #   
            if event.key == pygame.K_UP:
                angle = min(angle + 5, 80)
            elif event.key == pygame.K_DOWN:
                angle = max(angle - 5, 10)
            # 
            elif event.key == pygame.K_SPACE and on_ground:
                rad = math.radians(angle)
                vx = power * math.cos(rad)
                vy = -power * math.sin(rad)
                on_ground = False

    if not on_ground:
        vy += gravity
        x += vx
        y += vy

    if y + radius >= 400:
        y = 400 - radius
        vy = -vy * bounce_loss
        vx *= bounce_loss
        if abs(vy) < 1:
            vy = 0
            on_ground = True

    #    
    hoop_rect = pygame.Rect(hoop_x, hoop_y, hoop_width, hoop_height)
    ball_rect = pygame.Rect(x - radius, y - radius, radius * 2, radius * 2)
    scored = ball_rect.colliderect(hoop_rect)

    screen.fill((135, 206, 235))
    pygame.draw.rect(screen, (255, 100, 0), hoop_rect) 
    pygame.draw.circle(screen, (255, 165, 0), (int(x), int(y)), radius)  

    font = pygame.font.Font(None, 24)
    text = font.render(f": {angle}", True, (0, 0, 0))
    screen.blit(text, (10, 10))

    if scored:
        hit_text = font.render("!", True, (0, 255, 0))
        screen.blit(hit_text, (250, 50))

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import math
pygame.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("   ")

start_x, start_y = 100, 300
x, y = start_x, start_y
radius = 15

# 
vx, vy = 0, 0           #   
gravity = 0.5           # 
power = 12              #  
angle = 45              #    
bounce_loss = 0.7       #    
on_ground = True

hoop_x, hoop_y = 500, 200
hoop_width, hoop_height = 60, 10

font = pygame.font.Font(None, 24)

scored_flag = False

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                angle = min(angle + 5, 80)
            elif event.key == pygame.K_DOWN:
                angle = max(angle - 5, 10)
            elif event.key == pygame.K_w:
                power = min(power + 1, 25)
            elif event.key == pygame.K_s:
                power = max(power - 1, 5)
            elif event.key == pygame.K_SPACE and on_ground:
                rad = math.radians(angle)
                vx = power * math.cos(rad)
                vy = -power * math.sin(rad)
                on_ground = False

    if not on_ground:
        vy += gravity
        x += vx
        y += vy

    if y + radius >= 400:
        y = 400 - radius
        vy = -vy * bounce_loss
        vx *= bounce_loss
        if abs(vy) < 1 and abs(vx) < 1:
            x, y = start_x, start_y
            vx, vy = 0, 0
            on_ground = True
            scored_flag = False  

    hoop_rect = pygame.Rect(hoop_x, hoop_y, hoop_width, hoop_height)
    ball_rect = pygame.Rect(x - radius, y - radius, radius * 2, radius * 2)

    if ball_rect.colliderect(hoop_rect):
        scored_flag = True

    screen.fill((135, 206, 235))

    if on_ground:
        points = []
        rad = math.radians(angle)
        temp_vx = power * math.cos(rad)
        temp_vy = -power * math.sin(rad)
        temp_x, temp_y = start_x, start_y
        for i in range(60):
            temp_vy += gravity
            temp_x += temp_vx
            temp_y += temp_vy
            if temp_y + radius >= 400:
                break
            points.append((int(temp_x), int(temp_y)))
        if len(points) > 1:
            pygame.draw.lines(screen, (0, 0, 0), False, points, 2)

    pygame.draw.rect(screen, (255, 100, 0), hoop_rect)
    pygame.draw.circle(screen, (255, 165, 0), (int(x), int(y)), radius)

    text_angle = font.render(f": {angle}", True, (0, 0, 0))
    text_power = font.render(f": {power}", True, (0, 0, 0))
    screen.blit(text_angle, (10, 10))
    screen.blit(text_power, (10, 30))

    if scored_flag:
        hit_text = font.render("!", True, (0, 255, 0))
        screen.blit(hit_text, (250, 50))

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame
import math
pygame.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("   ")

background = pygame.image.load("basketball.jpg")
background = pygame.transform.scale(background, (600, 400))
ball_image = pygame.image.load("ball.jpg")
ball_image = pygame.transform.scale(ball_image, (30, 30))

pygame.mixer.init()
pygame.mixer.music.load("play.wav")
pygame.mixer.music.play(-1) 

score_sound = pygame.mixer.Sound("score.wav")  

start_x, start_y = 100, 300
x, y = start_x, start_y
radius = 15

vx, vy = 0, 0
gravity = 0.5
power = 12
angle = 45
bounce_loss = 0.7
on_ground = True

hoop_x, hoop_y = 500, 200
hoop_width, hoop_height = 60, 10

font = pygame.font.Font(None, 24)

scored_flag = False
score = 0

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                angle = min(angle + 5, 80)
            elif event.key == pygame.K_DOWN:
                angle = max(angle - 5, 10)
            elif event.key == pygame.K_w:
                power = min(power + 1, 25)
            elif event.key == pygame.K_s:
                power = max(power - 1, 5)
            elif event.key == pygame.K_SPACE and on_ground:
                rad = math.radians(angle)
                vx = power * math.cos(rad)
                vy = -power * math.sin(rad)
                on_ground = False

    if not on_ground:
        vy += gravity
        x += vx
        y += vy

    if y + radius >= 400:
        y = 400 - radius
        vy = -vy * bounce_loss
        vx *= bounce_loss
        if abs(vy) < 1 and abs(vx) < 1:
            x, y = start_x, start_y
            vx, vy = 0, 0
            on_ground = True
            scored_flag = False

    hoop_rect = pygame.Rect(hoop_x, hoop_y, hoop_width, hoop_height)
    ball_rect = pygame.Rect(x - radius, y - radius, radius * 2, radius * 2)

    if ball_rect.colliderect(hoop_rect) and not scored_flag:
        scored_flag = True
        score += 1
        score_sound.play()

    screen.blit(background, (0, 0))

    if on_ground:
        points = []
        rad = math.radians(angle)
        temp_vx = power * math.cos(rad)
        temp_vy = -power * math.sin(rad)
        temp_x, temp_y = start_x, start_y
        for i in range(60):
            temp_vy += gravity
            temp_x += temp_vx
            temp_y += temp_vy
            if temp_y + radius >= 400:
                break
            points.append((int(temp_x), int(temp_y)))
        if len(points) > 1:
            pygame.draw.lines(screen, (0, 0, 0), False, points, 2)

    pygame.draw.rect(screen, (255, 100, 0), hoop_rect)
    screen.blit(ball_image, (int(x - radius), int(y - radius)))

    text_angle = font.render(f": {angle}", True, (255, 0, 0))
    text_power = font.render(f": {power}", True, (255, 0, 0))
    text_score = font.render(f": {score}", True, (255, 0, 0))
    screen.blit(text_angle, (10, 10))
    screen.blit(text_power, (10, 30))
    screen.blit(text_score, (10, 50))

    if scored_flag:
        hit_text = font.render("!", True, (0, 255, 0))
        screen.blit(hit_text, (250, 50))

    pygame.display.flip()
    pygame.time.delay(20)

pygame.quit()
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("  ( )")

#   1
x1, y1 = 50, 50
vx1, vy1 = 1, 1
radius1 = 20
color1 = (255, 0, 0)

#   2
x2, y2 = 200, 150
vx2, vy2 = 1, 1
radius2 = 30
color2 = (0, 0, 255)

#  
def move(x, y, vx, vy, radius):
    x += vx
    y += vy
    if x - radius < 0 or x + radius > 400:
        vx = -vx
    if y - radius < 0 or y + radius > 300:
        vy = -vy
    return x, y, vx, vy

#  
def draw(x, y, radius, color):
    pygame.draw.circle(screen, color, (int(x), int(y)), radius)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    #   
    x1, y1, vx1, vy1 = move(x1, y1, vx1, vy1, radius1)
    x2, y2, vx2, vy2 = move(x2, y2, vx2, vy2, radius2)

    #  
    screen.fill((255, 255, 255))
    draw(x1, y1, radius1, color1)
    draw(x2, y2, radius2, color2)

    pygame.display.flip()
    pygame.time.delay(10)

pygame.quit()
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((400, 300))
pygame.display.set_caption("")

class Ball:
    def __init__(self, x, y, radius, color, vx, vy):
        self.x = x
        self.y = y
        self.radius = radius
        self.color = color
        self.vx = vx
        self.vy = vy

    #   
    def update(self):
        self.x += self.vx
        self.y += self.vy
        if self.x - self.radius < 0 or self.x + self.radius > 400:
            self.vx = -self.vx
        if self.y - self.radius < 0 or self.y + self.radius > 300:
            self.vy = -self.vy

    # 
    def draw(self):
        pygame.draw.circle(screen, self.color, (int(self.x), int(self.y)), self.radius)

#   
balls = [
    Ball(50, 50, 20, (255,0,0), 3, 2),
    Ball(150, 100, 30, (0,0,255), 2, 3)  #    
]

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((255, 255, 255))
    for ball in balls:
        ball.update()
        ball.draw()
    pygame.display.flip()
    pygame.time.delay(10)

pygame.quit()
```
