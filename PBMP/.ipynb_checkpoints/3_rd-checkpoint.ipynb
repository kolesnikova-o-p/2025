```python
import turtle

t = turtle.Turtle()
t.shape("turtle")
t.speed(3)
t.color("green")

def parallelogram(size=100):
    for i in range(2):
        t.forward(size)
        t.right(60)
        t.forward(size)
        t.right(120)

for i in range(3):
    parallelogram(100)
    t.penup()      # поднять перо, чтобы не оставлять линию
    t.forward(150) # смещение вправо
    t.pendown()    # опустить перо

turtle.done()
```

```python
import turtle

t = turtle.Turtle()
t.speed(5)
turtle.bgcolor("skyblue")

t.penup()
t.goto(-100, -100)  
t.pendown()
t.color("brown")
t.begin_fill()
for i in range(2):
    t.forward(200)  
    t.left(90)
    t.forward(150)   
    t.left(90)
t.end_fill()

t.penup()
t.goto(-40, -40)
t.pendown()
t.color("yellow")
t.begin_fill()
for i in range(4):
    t.forward(80)
    t.left(90)
t.end_fill()


t.penup()
t.goto(-100, 50) 
t.pendown()
t.color("peru")
t.begin_fill()
t.goto(0, 150)     # вершина крыши (по центру)
t.goto(100, 50)    # правая точка крыши
t.goto(-100, 50)   # возвращаемся влево
t.end_fill()

t.penup()
t.goto(40, 80)   
t.pendown()
t.color("sienna")
t.begin_fill()
for i in range(2):
    t.forward(30)
    t.left(90)
    t.forward(60)
    t.left(90)
t.end_fill()


t.penup()
t.goto(55, 140)  
t.pendown()
t.color("gray")
for r in [15, 20, 25]:
    t.penup()
    t.goto(55, 140 + r * 2) 
    t.pendown()
    t.begin_fill()
    t.circle(r)
    t.end_fill()

turtle.done()
```

```python
from random import randint
import turtle

screen = turtle.Screen()
screen.setup(800, 600)

number_of_turtles = 5
steps_of_time_number = 100

pool = [turtle.Turtle(shape='turtle') for i in range(number_of_turtles)]
# цикл по всем черепашкам в списке
for unit in pool:
    unit.penup()                # поднимаем "перо", чтобы не рисовать линию при движении
    unit.speed(50)              # ставим высокую скорость движения (чем больше число, тем быстрее)
    unit.goto(randint(-200, 200), randint(-200, 200))  # перемещаем черепашку в случайное место на экране

# запускаем движение на указанное количество шагов
for i in range(steps_of_time_number):
    for unit in pool:
        unit.forward(2)         # двигаем каждую черепашку вперёд на 2 единицы каждый шаг!

turtle.done()
```

```python
from random import randint, choice
import turtle

screen = turtle.Screen()
screen.setup(800, 600)

# список файлов с картинками (должны быть .gif в папке со скриптом)
bird_images = ["bird1.gif", "bird2.gif", "bird3.gif", "bird4.gif", "bird5.gif"]

# регистрируем каждую картинку как новую форму
for img in bird_images:
    screen.addshape(img)

number_of_turtles = 5

pool = [turtle.Turtle(shape=bird_images[i]) for i in range(number_of_turtles)]

for unit in pool:
    unit.penup()
    unit.speed(0) 
    unit.goto(randint(-200, 200), randint(-200, 200))
    unit.setheading(randint(0, 360))  # случайный угол

def move():
    # двигаем всех птичек
    for unit in pool:
        unit.forward(2)
    # повторяем каждые 50 мс
    screen.ontimer(move, 50) #это чтобы если через 50 мс окно закрыто, то все!

move()
turtle.done()
```

turtle.done() нужно для того, чтобы окно  не закрылось сразу после запуска и продолжало работать, пока не будет закрыто

```python
from random import randint, choice
import turtle
import time

screen = turtle.Screen()
screen.setup(800, 600)

bird_images = ["bird1.gif", "bird2.gif", "bird3.gif", "bird4.gif", "bird5.gif"]

for img in bird_images:
    screen.addshape(img)

number_of_turtles = 5
steps_of_time_number = 300

pool = [turtle.Turtle(shape=bird_images[i]) for i in range(number_of_turtles)]


for unit in pool:
    unit.penup()
    unit.speed(0)  
    unit.goto(randint(-300, 300), randint(-200, 200))  
    unit.setheading(randint(0, 360)) 

for i in range(steps_of_time_number):
    for unit in pool:
        unit.forward(5)  # движение вперёд
    time.sleep(0.05)  # задержка для плавности

turtle.done()
```

## Полезные примеры с черепашкой

### Идеальный газ

```python
from random import randint
import turtle
import math

number_of_turtles = 15
steps_of_time_number = 10000

# Настройка экрана
turtle.tracer(False)
screen = turtle.Screen()
screen.setup(500, 500)

# Создаём черепашек в случайных местах и с случайным направлением
pool = []
for i in range(number_of_turtles):
    t = turtle.Turtle(shape='turtle')
    t.penup()
    t.speed(0)
    t.goto(randint(-200, 200), randint(-200, 200))
    t.setheading(randint(0, 360))  # случайное направление
    pool.append(t)

# Параметры «сосуда»
border = 230   # от центра до стенок
speed = 2      # шаг за итерацию

for i in range(steps_of_time_number):
    for unit in pool:
        unit.forward(speed)
        x, y = unit.position()
        heading = unit.heading()

        # Отражение от вертикальных стен
        if abs(x) > border:
            unit.setheading(180 - heading)
            unit.setx(math.copysign(border, x))

        # Отражение от горизонтальных стен
        if abs(y) > border:
            unit.setheading(-heading)
            unit.sety(math.copysign(border, y))

    turtle.update()

turtle.done()
```

```python
import turtle

def polygon(t, n, size):
    for i in range(n):
        t.forward(size)
        t.left(360/n)

screen = turtle.Screen()
screen.bgcolor("lightblue")

screen.addshape("flower.gif") 

t = turtle.Turtle()
t.shape("flower.gif")  
t.speed(0)

for i in range(36):
    polygon(t, 6, 100)
    t.stamp()       
    t.left(10)

turtle.done()
```

```python
from PIL import Image
img = Image.open("flower.gif")

scales = {
    "small": 0.25,
    "medium": 0.5,
    "large": 0.75
}

for name, scale in scales.items():
    new_size = (int(img.width * scale), int(img.height * scale))
    resized = img.resize(new_size, Image.LANCZOS)
    filename = f"flower_{name}.gif"
    resized.save(filename)
    print(f"Сохранено: {filename} ({new_size[0]}x{new_size[1]})")
```

```python
import turtle
import random

def polygon(t, n, size):
    for i in range(n):
        t.forward(size)
        t.left(360/n)

# экран + регистрация форм
screen = turtle.Screen()
flowers = ["flower_small.gif", "flower_medium.gif", "flower_large.gif"]

for f in flowers:
    screen.addshape(f)

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.penup()

for i in range(36):
    polygon(t, 6, 100)
    t.left(10)

    if i % 6 == 0:  # каждые 6 шагов
        flower.shape(random.choice(flowers))
        flower.goto(t.xcor(), t.ycor())
        flower.stamp()

turtle.done()
```

```python
import turtle

def polygon_with_stamps(t, n, size, stamp_turtle):
    angle = 360 / n
    for i in range(n):
        t.forward(size)
        stamp_turtle.goto(t.xcor(), t.ycor())
        stamp_turtle.stamp()
        t.left(angle)

screen = turtle.Screen()
screen.addshape("flower_small.gif")

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.shape("flower_small.gif")
flower.penup()

for i in range(36):
    polygon_with_stamps(t, 6, 100, flower)
    t.left(10)

turtle.done()
```

```python
import turtle

def polygon_third_vertex_stamp(t, n, size, stamp_turtle):
    angle = 360 / n
    for i in range(n):
        t.forward(size)
        t.left(angle)
        if i == 2:  # третья вершина
            stamp_turtle.goto(t.xcor(), t.ycor())
            stamp_turtle.stamp()

screen = turtle.Screen()
screen.addshape("flower_small.gif")

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.shape("flower_small.gif")
flower.penup()

for i in range(36):
    polygon_third_vertex_stamp(t, 6, 100, flower)
    t.left(10)

turtle.done()
```

# PyGame

pygame — это популярный кроссплатформенный модуль для языка Python, предназначенный для разработки 2D-игр и мультимедийных приложений. Он обёрнут вокруг низкоуровневой библиотеки SDL (Simple DirectMedia Layer) и предоставляет удобный, «питонический» интерфейс для работы с графикой, звуком, событиями и вводом.

Основные возможности

Создание окон, отрисовка спрайтов, работа с изображениями (PNG, JPG, GIF и др.).

Поддержка прозрачности, альфа-каналов, вращения, масштабирования.

Управление кадрами через объект pygame.time.Clock для плавной анимации.

Обработка ввода

События клавиатуры, мыши, геймпадов и сенсорных устройств.

Гибкая система очередей событий (pygame.event).

Воспроизведение звуковых эффектов (pygame.mixer.Sound) и музыки (pygame.mixer.music) в разных форматах.

Работа с текстом

Рендеринг шрифтов (через pygame.font) для вывода текста на экран.

```python
import pygame

# После импорта библиотеки, необходимо её инициализировать:
pygame.init()

# И создать окно:
screen = pygame.display.set_mode((300, 200))

# здесь будут рисоваться фигуры
# ...

# после чего, чтобы они отобразились на экране, экран нужно обновить:
pygame.display.update()
# Эту же команду нужно будет повторять, если на экране происходят изменения.

# Наконец, нужно создать основной цикл, в котором будут отслеживаться
# происходящие события.
# Пока единственное событие, которое нас интересует - выход из программы.
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
```

```python
import pygame

pygame.init()                          # инициализация всех модулей
screen = pygame.display.set_mode((400, 300))  # создаём окно 400×300
pygame.display.set_caption("Самый простой пример")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:   # пользователь нажал "закрыть"
            running = False

    screen.fill((30, 144, 255))         # закрашиваем фон синим цветом

pygame.quit()                           # корректно выходим
```

Аргумент цвета может быть:

тройкой (R, G, B) — значения красного, зелёного и синего каналов (0–255)

четвёркой (R, G, B, A) — если нужен альфа-канал (прозрачность)

```python
pygame.init()                         
screen = pygame.display.set_mode((400, 300))  
pygame.display.set_caption("Самый простой пример")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:  
            running = False

    screen.fill((30, 144, 255))         
    pygame.display.flip()               # обновляем экран

pygame.quit()                        
```

ВАЖНО!

Что делает:
Переключает весь экран сразу — показывает весь новый кадр, который вы нарисовали на поверхности screen.

### Смайл

```python
import pygame

pygame.init()                         
screen = pygame.display.set_mode((400, 300))  
pygame.display.set_caption("Весёлый смайлик")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:  
            running = False

    screen.fill((255, 255, 255)) 

    pygame.draw.circle(screen, (255, 230, 0), (200, 150), 100)    
    pygame.draw.circle(screen, (0, 0, 0),   (200, 150), 100, 3)     

    pygame.draw.circle(screen, (0, 0, 0), (160, 120), 15)
    pygame.draw.circle(screen, (0, 0, 0), (240, 120), 15)

    smile_rect = pygame.Rect(130, 120, 140, 100)
    pygame.draw.arc(screen, (0, 0, 0), smile_rect, 3.14, 2*3.14, 4)

    pygame.display.flip()             

pygame.quit()
```
