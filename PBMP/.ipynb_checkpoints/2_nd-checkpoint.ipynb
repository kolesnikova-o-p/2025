```python
def average(lst):
    if len(lst) != 0:
        return sum(lst)/len(lst)
    else:
        return None
print("Среднее:", average([1,2,3,4,5]))
```

```python
def unique(lst):
    return list(set(lst))

print("Уникальные:", unique([1,2,2,3,4,4,5]))
```

```python
def factorial(n):
     if n==0:
         return 1
     else:
         return n*factorial(n-1)

print("Факториал 5:", factorial(5))
```

```python
import turtle

t = turtle.Turtle()
t.shape("turtle")
t.speed(3)
t.color("green")

def parallelogram(size=100):
    for i in range(2):
        t.forward(size)
        t.right(60)
        t.forward(size)
        t.right(120)

for i in range(3):
    parallelogram(100)
    t.penup()      # поднять перо, чтобы не оставлять линию
    t.forward(150) # смещение вправо
    t.pendown()    # опустить перо

turtle.done()
```

```python
import turtle

t = turtle.Turtle()
t.speed(5)
turtle.bgcolor("skyblue")

t.penup()
t.goto(-100, -100)  
t.pendown()
t.color("brown")
t.begin_fill()
for i in range(2):
    t.forward(200)  
    t.left(90)
    t.forward(150)   
    t.left(90)
t.end_fill()

t.penup()
t.goto(-40, -40)
t.pendown()
t.color("yellow")
t.begin_fill()
for i in range(4):
    t.forward(80)
    t.left(90)
t.end_fill()


t.penup()
t.goto(-100, 50) 
t.pendown()
t.color("peru")
t.begin_fill()
t.goto(0, 150)     # вершина крыши (по центру)
t.goto(100, 50)    # правая точка крыши
t.goto(-100, 50)   # возвращаемся влево
t.end_fill()

t.penup()
t.goto(40, 80)   
t.pendown()
t.color("sienna")
t.begin_fill()
for i in range(2):
    t.forward(30)
    t.left(90)
    t.forward(60)
    t.left(90)
t.end_fill()


t.penup()
t.goto(55, 140)  
t.pendown()
t.color("gray")
for r in [15, 20, 25]:
    t.penup()
    t.goto(55, 140 + r * 2) 
    t.pendown()
    t.begin_fill()
    t.circle(r)
    t.end_fill()

turtle.done()
```

```python
from random import randint
import turtle

screen = turtle.Screen()
screen.setup(800, 600)

number_of_turtles = 5
steps_of_time_number = 100

pool = [turtle.Turtle(shape='turtle') for i in range(number_of_turtles)]
# цикл по всем черепашкам в списке
for unit in pool:
    unit.penup()                # поднимаем "перо", чтобы не рисовать линию при движении
    unit.speed(50)              # ставим высокую скорость движения (чем больше число, тем быстрее)
    unit.goto(randint(-200, 200), randint(-200, 200))  # перемещаем черепашку в случайное место на экране

# запускаем движение на указанное количество шагов
for i in range(steps_of_time_number):
    for unit in pool:
        unit.forward(2)         # двигаем каждую черепашку вперёд на 2 единицы каждый шаг!

turtle.done()
```

```python
from random import randint, choice
import turtle

screen = turtle.Screen()
screen.setup(800, 600)

# список файлов с картинками (должны быть .gif в папке со скриптом)
bird_images = ["bird1.gif", "bird2.gif", "bird3.gif", "bird4.gif", "bird5.gif"]

# регистрируем каждую картинку как новую форму
for img in bird_images:
    screen.addshape(img)

number_of_turtles = 5

pool = [turtle.Turtle(shape=bird_images[i]) for i in range(number_of_turtles)]

for unit in pool:
    unit.penup()
    unit.speed(0) 
    unit.goto(randint(-200, 200), randint(-200, 200))
    unit.setheading(randint(0, 360))  # случайный угол

def move():
    # двигаем всех птичек
    for unit in pool:
        unit.forward(2)
    # повторяем каждые 50 мс
    screen.ontimer(move, 50) #это чтобы если через 50 мс окно закрыто, то все!

move()
turtle.done()
```

turtle.done() нужно для того, чтобы окно  не закрылось сразу после запуска и продолжало работать, пока не будет закрыто

```python
from random import randint, choice
import turtle
import time

screen = turtle.Screen()
screen.setup(800, 600)

bird_images = ["bird1.gif", "bird2.gif", "bird3.gif", "bird4.gif", "bird5.gif"]

for img in bird_images:
    screen.addshape(img)

number_of_turtles = 5
steps_of_time_number = 300

pool = [turtle.Turtle(shape=bird_images[i]) for i in range(number_of_turtles)]


for unit in pool:
    unit.penup()
    unit.speed(0)  
    unit.goto(randint(-300, 300), randint(-200, 200))  
    unit.setheading(randint(0, 360)) 

for i in range(steps_of_time_number):
    for unit in pool:
        unit.forward(5)  # движение вперёд
    time.sleep(0.05)  # задержка для плавности

turtle.done()
```

### Фрактальное дерево

```python
import turtle

def draw_branch(t, branch_length):
    if branch_length > 5:
        t.forward(branch_length)
        t.right(20)
        draw_branch(t, branch_length - 15)
        t.left(40)
        draw_branch(t, branch_length - 15)
        t.right(20)
        t.backward(branch_length)

screen = turtle.Screen()
t = turtle.Turtle()
t.left(90)
t.speed(10)
draw_branch(t, 100)
turtle.done()
```

### Спираль Фибоначчи

Числа Фибоначчи — это математическая последовательность, в которой каждое последующее число равно сумме двух предыдущих.

```python
import turtle
import math

fib = [0, 1]
for i in range(2, 10):
    fib.append(fib[-1] + fib[-2])

t = turtle.Turtle()
t.speed(10)
for i in range(2, 10):
    t.circle(fib[i]*10, 90)
turtle.done()
```

### Снежинка Коха

```python
import turtle

def koch(t, order, size):
    if order == 0:
        t.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
            koch(t, order-1, size/3)
            t.left(angle)

t = turtle.Turtle()
t.speed(0)
for i in range(3):
    koch(t, 4, 300)
    t.right(120)
turtle.done()
```

```python
from PIL import Image
img = Image.open("flower.gif")

scales = {
    "small": 0.25,
    "medium": 0.5,
    "large": 0.75
}

for name, scale in scales.items():
    new_size = (int(img.width * scale), int(img.height * scale))
    resized = img.resize(new_size, Image.LANCZOS)
    filename = f"flower_{name}.gif"
    resized.save(filename)
    print(f"Сохранено: {filename} ({new_size[0]}x{new_size[1]})")
```

```python
import turtle
import random

def polygon(t, n, size):
    for i in range(n):
        t.forward(size)
        t.left(360/n)

# экран + регистрация форм
screen = turtle.Screen()
flowers = ["flower_small.gif", "flower_medium.gif", "flower_large.gif"]

for f in flowers:
    screen.addshape(f)

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.penup()

for i in range(36):
    polygon(t, 6, 100)
    t.left(10)

    if i % 6 == 0:  # каждые 6 шагов
        flower.shape(random.choice(flowers))
        flower.goto(t.xcor(), t.ycor())
        flower.stamp()

turtle.done()
```

```python
import turtle

def polygon_with_stamps(t, n, size, stamp_turtle):
    angle = 360 / n
    for i in range(n):
        t.forward(size)
        stamp_turtle.goto(t.xcor(), t.ycor())
        stamp_turtle.stamp()
        t.left(angle)

screen = turtle.Screen()
screen.addshape("flower_small.gif")

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.shape("flower_small.gif")
flower.penup()

for i in range(36):
    polygon_with_stamps(t, 6, 100, flower)
    t.left(10)

turtle.done()
```

```python
import turtle

def polygon_third_vertex_stamp(t, n, size, stamp_turtle):
    angle = 360 / n
    for i in range(n):
        t.forward(size)
        t.left(angle)
        if i == 2:  # третья вершина
            stamp_turtle.goto(t.xcor(), t.ycor())
            stamp_turtle.stamp()

screen = turtle.Screen()
screen.addshape("flower_small.gif")

t = turtle.Turtle()
t.speed(0)

flower = turtle.Turtle()
flower.shape("flower_small.gif")
flower.penup()

for i in range(36):
    polygon_third_vertex_stamp(t, 6, 100, flower)
    t.left(10)

turtle.done()
```

### Игра крестики-нолики

```python
import turtle

def draw_grid():
    t = turtle.Turtle()
    t.speed(0)
    t.penup()
    for i in range(-1, 2):
        t.goto(-150, 50*i)
        t.pendown()
        t.forward(300)
        t.penup()
    t.left(90)
    for i in range(-1, 2):
        t.goto(50*i, -150)
        t.pendown()
        t.forward(300)
        t.penup()

def draw_x(x, y):
    t = turtle.Turtle()
    t.speed(0)
    t.penup()
    t.goto(x-40, y-40)
    t.pendown()
    t.goto(x+40, y+40)
    t.penup()
    t.goto(x-40, y+40)
    t.pendown()
    t.goto(x+40, y-40)

def draw_o(x, y):
    t = turtle.Turtle()
    t.speed(0)
    t.penup()
    t.goto(x, y-40)
    t.pendown()
    t.circle(40)

draw_grid()
draw_x(-100, 100)
draw_o(0, 100)
draw_x(100, -100)
turtle.done()
```
